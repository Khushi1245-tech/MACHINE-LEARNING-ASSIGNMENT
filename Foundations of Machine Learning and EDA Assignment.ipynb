## Question 2: Explain overfitting and underfitting in ML. How can you detect and prevent them? Hint: Discuss bias-variance tradeoff, cross-validation, and regularization techniques.

### Answer: **Overfitting vs Underfitting in ML**

**Overfitting**

* **Meaning:** The model learns training data *too well*, including noise and random patterns.
* **Symptom:** Very high training accuracy but poor test/validation accuracy.
* **Cause:** Model is too complex → low bias, high variance.

**Underfitting**

* **Meaning:** The model is too simple to capture the underlying pattern.
* **Symptom:** Low accuracy on both training and test data.
* **Cause:** Model is too simple → high bias, low variance.

### **How to Detect Them**

* **Overfitting:**

  * Large gap between training and validation error.
  * Validation loss increases while training loss decreases.

* **Underfitting:**

  * Both training and validation errors are high.
  * Model fails to improve even after training.


### **How to Prevent Them**

**For Overfitting:**

* Use **cross-validation** (e.g., k-fold) to check model performance.
* Apply **regularization** (L1, L2, dropout).
* Reduce model complexity (prune trees, fewer layers).
* Add more data or use data augmentation.
* Use **early stopping**.

**For Underfitting:**

* Increase model complexity (more features, deeper model).
* Reduce regularization.
* Train for more epochs.
* Improve feature engineering.

### **Bias–Variance Trade-off (in one line)**

* **Underfitting = high bias, low variance.**
* **Overfitting = low bias, high variance.**
* Goal: balance both for optimal performance.

## Question 3:How would you handle missing values in a dataset? Explain at least three methods with examples.
Hint: Consider deletion, mean/median imputation, and predictive modeling.

### Answer: **Handling Missing Values in a Dataset**

Missing values can distort analysis, reduce model accuracy, and lead to biased results. They must be treated carefully depending on the size, type, and pattern of missing data.

### **1. Deletion Methods**

#### **a) Listwise Deletion (Removing Rows)**

* Remove rows that contain missing values.
* **Example:** If a dataset has 2% rows with `Age = NaN`, delete those rows.
* **Use when:**

  * Missing data is very small and random (MCAR).
  * Does not affect sample size significantly.

#### **b) Column Deletion**

* Remove columns with a very high percentage of missing data.
* **Example:** A column with 85% missing values is usually dropped.
* **Use when:**

  * The feature is not critical or too incomplete to be useful.

### **2. Simple Imputation Methods**

#### **a) Mean/Median Imputation (Numerical Data)**

* Replace missing values using:

  * **Mean** for normally distributed data.
  * **Median** for skewed data.
* **Example:**

  * Salary column: [50k, 60k, **NaN**, 70k] → Median = 60k → Fill NaN with 60k.

#### **b) Mode Imputation (Categorical Data)**

* Replace NaN with the most frequent category.
* **Example:**

  * City column: [Delhi, Mumbai, **NaN**, Delhi] → Mode = Delhi → Fill NaN.

### **3. Advanced Imputation (Predictive Modeling)**

#### **a) KNN Imputation**

* Uses similarity between rows to estimate the missing value.
* Good when the dataset has meaningful patterns.

#### **b) Regression Imputation**

* Build a regression model to predict the missing values using other features.
* **Example:** Use salary, education, and experience to predict missing age values.

#### **c) MICE (Multiple Imputation by Chained Equations)**

* Creates multiple models and averages the results → more robust.
* Best for large datasets with complex relationships.

### **How to Choose the Right Method**

* **<5% missing values:** Use deletion or simple imputation.
* **Numeric skewed data:** Median imputation is better.
* **Categorical data:** Mode imputation.
* **Large datasets with strong correlations:** Predictive models or MICE.
* **ML tasks:** Try **KNN Imputer** or **Iterative Imputer** for better accuracy.

## Question 5: Why is feature scaling important in ML? Compare Min-Max scaling and Standardization.
Hint: Explain impact on distance-based algorithms (e.g., KNN, SVM) and gradient
descent.

### Answer: Feature scaling ensures that all features contribute equally to the model.
Without scaling, features with large numeric ranges dominate those with small ranges.

### **Why it matters:**

* **Distance-based algorithms (KNN, K-Means, SVM):**
  These rely on Euclidean distance. If one feature is in thousands and another in decimals, the larger one will dominate → wrong predictions.
* **Gradient Descent (Linear/Logistic Regression, Neural Networks):**
  Unscaled data slows down convergence, causes zig-zag movement of gradients, and may fail to reach the global minimum.
* **Faster training + better accuracy:**
  Models become stable, faster, and more accurate.

## **Min-Max Scaling vs Standardization**

### **1. Min-Max Scaling (Normalization)**

* **Formula:**
  [
  x' = \frac{x - \min(x)}{\max(x) - \min(x)}
  ]
* **Range:** 0 to 1
* **Best For:**

  * Algorithms that require bounded values (KNN, Neural Networks, SVM with RBF).
  * When data is not normally distributed.
* **Issue:**
  Very sensitive to outliers.


### **2. Standardization (Z-Score Scaling)**

* **Formula:**
  [
  x' = \frac{x - \mu}{\sigma}
  ]
* **Range:** Mean = 0, Std = 1 (not bounded)
* **Best For:**

  * Algorithms that assume Gaussian distribution (Linear Regression, Logistic Regression).
  * Gradient-based models (SGD, Neural Networks).
  * Less sensitive to outliers than Min-Max.


## **Quick Comparison Table**

| Method              | Output Range  | Best For                | Sensitive to Outliers | Use Cases               |
| ------------------- | ------------- | ----------------------- | --------------------- | ----------------------- |
| **Min-Max Scaling** | 0 to 1        | KNN, K-Means, SVM (RBF) | High                  | Distance-based models   |
| **Standardization** | Mean 0, Std 1 | Regression, SVM, NN     | Low                   | Gradient descent models |

## Question 6: Compare Label Encoding and One-Hot Encoding. When would you prefer one over the other?
Hint: Consider categorical variables with ordinal vs. nominal relationships.

### Answer:**Label Encoding vs. One-Hot Encoding**

### **1. Label Encoding**

* **What it does:** Converts each category into an **integer number** (0,1,2,…).
* **Example:**

  * Size = {Small → 0, Medium → 1, Large → 2}
* **Good For:**

  * **Ordinal** categories (where order matters).
  * Example: Size, Rating (Low < Medium < High), Education Level.
* **Problem:**

  * Creates a **false numeric relationship** if used on nominal data.
  * Model may assume *Large > Small* (which is wrong for nominal values like cities).

### **2. One-Hot Encoding**

* **What it does:** Creates **separate binary columns** for each category.
* **Example:**

  * City = Delhi → [1,0,0]
  * Mumbai → [0,1,0]
* **Good For:**

  * **Nominal** categories (no natural order).
  * Example: Gender, City, Color, Product Type.
* **Advantages:**

  * No ordinal relationship assumed.
* **Drawback:**

  * Increases number of features (**high dimensionality**) for large category sets.

## **When to Use Which?**

###  **Use Label Encoding When:**

* The categories have a **natural ranking**.
* Examples:

  * Education Level (10th < 12th < Graduate)
  * Size (S < M < L)
  * Customer rating (1–5)

###  **Use One-Hot Encoding When:**

* Categories have **no order** (nominal data).
* Examples:

  * City
  * Color
  * Country
  * Gender
* Also preferred in **tree-based models** (Decision Trees, Random Forests).

## **Quick Comparison Table**

| Aspect           | Label Encoding                      | One-Hot Encoding    |
| ---------------- | ----------------------------------- | ------------------- |
| Output           | Integer values                      | Binary columns      |
| Best For         | Ordinal data                        | Nominal data        |
| Preserves Order? | Yes                                 | No                  |
| Risk             | False ranking on nominal categories | High dimensionality |
| Used In          | Trees, ordinal data                 | KNN, LR, SVM, NN    |
