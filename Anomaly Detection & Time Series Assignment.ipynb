## Question 1: What is Anomaly Detection? Explain its types (point, contextual, and collective anomalies) with examples.


### Answer:**Anomaly Detection** is a data analysis technique used to identify unusual, rare, or abnormal patterns in data that significantly deviate from normal behavior. These anomalies often indicate important events such as fraud, system faults, cyber-attacks, medical abnormalities, or data errors. Anomaly detection is broadly classified into three types: **point anomalies**, **contextual anomalies**, and **collective anomalies**. A **point anomaly** occurs when a single data point is clearly different from the rest of the dataset, such as an unusually large bank transaction (₹5,00,000) compared to normal daily transactions of a few thousand rupees. A **contextual anomaly** occurs when a data point is normal in general but becomes abnormal in a specific situation or context, for example, a temperature of 30°C is normal in summer but abnormal in winter, or a login at 2 AM may be normal for a night-shift worker but suspicious for a regular office employee. A **collective anomaly** occurs when a group of data points together form an abnormal pattern even though individual values may appear normal, such as continuous irregular heartbeats in ECG data or a sustained unusual traffic flow in a network indicating a cyber-attack. Together, these types help in detecting hidden risks and unusual behaviors across different real-world systems.

## Question 2: Compare Isolation Forest, DBSCAN, and Local Outlier Factor in terms of their approach and suitable use cases.

### Answer: **Isolation Forest, DBSCAN, and Local Outlier Factor (LOF)** are popular unsupervised anomaly detection techniques, but they differ in their underlying approach and suitable use cases. **Isolation Forest** works by randomly partitioning the data using decision-tree-like structures and isolates anomalies faster than normal points because anomalies are few and different; it is highly efficient on large, high-dimensional datasets and is widely used in fraud detection, network intrusion detection, and large-scale monitoring systems. **DBSCAN** (Density-Based Spatial Clustering of Applications with Noise) is a density-based clustering algorithm that identifies anomalies as points that do not belong to any dense cluster; it is best suited for spatial data, geospatial analysis, and datasets where clusters of arbitrary shapes exist, such as location-based services and image segmentation. **Local Outlier Factor (LOF)**, on the other hand, detects anomalies by comparing the local density of a point with the density of its neighbors, labeling points as outliers if they lie in regions of significantly lower density; it is especially useful in datasets with varying densities, such as network traffic data, sensor networks, and behavioral data analysis. In summary, Isolation Forest is ideal for large-scale, high-dimensional data, DBSCAN is best for density-based spatial anomaly detection, and LOF is most suitable for detecting local anomalies in complex, uneven-density datasets.

## Question 3: What are the key components of a Time Series? Explain each with one example.

### Answer:**A time series** is a sequence of data points collected or recorded at regular time intervals over time (such as daily, monthly, or yearly). The key components of a time series are **trend, seasonality, cyclical variation, and irregular (random) variation**, and each component represents a different type of pattern in the data. The **trend** refers to the long-term upward or downward movement in the data, such as a steady increase in annual company revenue over several years. **Seasonality** represents regular, repeating patterns that occur at fixed time intervals due to seasonal factors, for example, higher electricity consumption every summer or increased retail sales during festivals each year. **Cyclical variation** refers to long-term fluctuations that do not follow a fixed period and are usually linked to economic or business cycles, such as rises and falls in the stock market over several years. The **irregular (random) component** represents unpredictable, random variations caused by unexpected events, such as a sudden drop in tourism due to a natural disaster or pandemic. Together, these components help in understanding, analyzing, and forecasting time-series data effectively.

## Question 4: Define Stationary in time series. How can you test and transform a non-stationary series into a stationary one?

### Answer:**In time series analysis, a series is said to be *stationary* if its statistical properties—such as mean, variance, and autocorrelation—remain constant over time.** In other words, a stationary series does not exhibit trends, seasonality, or other systematic changes over time. Stationarity is important because many time series forecasting models, like ARIMA, assume that the underlying data is stationary.

To **test whether a time series is stationary**, common methods include:

1. **Visual Inspection:** Plot the series to check for trends or changing variance.
2. **Statistical Tests:**

   * **Augmented Dickey-Fuller (ADF) test:** Checks for the presence of a unit root; a low p-value (< 0.05) indicates stationarity.
   * **KPSS test:** Checks the null hypothesis of stationarity; a high p-value (> 0.05) indicates stationarity.

To **transform a non-stationary series into a stationary one**, several techniques can be applied:

1. **Differencing:** Subtract the previous observation from the current one to remove trends. Example: ( y'*t = y_t - y*{t-1} )
2. **Log Transformation:** Apply log to stabilize variance, especially for exponential growth.
3. **Detrending:** Remove the trend component using regression or moving averages.
4. **Seasonal Differencing:** Subtract the value from the same period in the previous season to remove seasonality.

By applying these methods, a non-stationary series can be converted into a stationary one, making it suitable for time series modeling and accurate forecasting.

## Question 5: Differentiate between AR, MA, ARIMA, SARIMA, and SARIMAX models in terms of structure and application.

#### Answer: **AR, MA, ARIMA, SARIMA, and SARIMAX** are time series forecasting models that differ in structure and application. **AR (AutoRegressive) model** predicts the current value of a series based on its own previous values (lags), capturing temporal correlation; for example, predicting today’s temperature using temperatures of the past few days. **MA (Moving Average) model** predicts the current value based on past forecast errors, smoothing out random shocks, such as modeling stock price fluctuations influenced by recent unexpected market changes. **ARIMA (AutoRegressive Integrated Moving Average)** combines both AR and MA components along with differencing (integration) to handle non-stationary data, making it suitable for many general time series forecasting tasks like sales or demand prediction. **SARIMA (Seasonal ARIMA)** extends ARIMA by adding seasonal terms to account for repeating patterns at fixed intervals, such as monthly retail sales with yearly seasonal effects. **SARIMAX (Seasonal ARIMA with eXogenous variables)** further extends SARIMA by including external variables (exogenous factors) that may influence the series, like incorporating advertising spend or holidays when forecasting product demand. In short, AR and MA handle stationary series with simple structures, ARIMA handles non-stationary series, SARIMA adds seasonality, and SARIMAX adds external influences for more accurate forecasting.

## Question 6: Load a time series dataset (e.g., AirPassengers), plot the original series, and decompose it into trend, seasonality, and residual components
(Include your Python code and output in the code box below.)


```python
import pandas as pd
import matplotlib.pyplot as plt
from statsmodels.tsa.seasonal import seasonal_decompose

# 1. Load the AirPassengers dataset
url = "https://raw.githubusercontent.com/jbrownlee/Datasets/master/airline-passengers.csv"
data = pd.read_csv(url, parse_dates=['Month'], index_col='Month')

# 2. Plot the original time series
plt.figure(figsize=(10, 6))
plt.plot(data, label='Original Series')
plt.title('Monthly Airline Passengers (1949-1960)')
plt.xlabel('Year')
plt.ylabel('Passengers')
plt.legend()
plt.show()

# 3. Decompose into Trend, Seasonality, and Residuals
# We use 'multiplicative' as the seasonal variance increases with the trend
result = seasonal_decompose(data['Passengers'], model='multiplicative', period=12)

# 4. Plot the decomposed components
result.plot()
plt.show()
```

## Question 7: Apply Isolation Forest on a numerical dataset (e.g., NYC Taxi Fare) to detect anomalies. Visualize the anomalies on a 2D scatter plot.
(Include your Python code and output in the code box below.)

```python
# Import libraries
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
from sklearn.ensemble import IsolationForest

# Generate synthetic NYC Taxi fare dataset
np.random.seed(42)
n_samples = 1000

# Normal rides
fare_amount = np.random.normal(loc=15, scale=5, size=n_samples)  # average fare $15
passenger_count = np.random.randint(1, 5, size=n_samples)        # 1-4 passengers

# Create anomalies
fare_amount = np.append(fare_amount, [200, 300, 500])  # extremely high fares
passenger_count = np.append(passenger_count, [1, 6, 8])  # unusual passenger counts

# Combine into DataFrame
data = pd.DataFrame({'fare_amount': fare_amount, 'passenger_count': passenger_count})

# Features for Isolation Forest
X = data[['fare_amount', 'passenger_count']]

# Initialize Isolation Forest
iso_forest = IsolationForest(contamination=0.01, random_state=42)
data['anomaly'] = iso_forest.fit_predict(X)

# Separate anomalies and normal points
anomalies = data[data['anomaly'] == -1]
normal = data[data['anomaly'] == 1]

# Plot results
plt.figure(figsize=(10,6))
plt.scatter(normal['fare_amount'], normal['passenger_count'], c='blue', label='Normal', alpha=0.5)
plt.scatter(anomalies['fare_amount'], anomalies['passenger_count'], c='red', label='Anomaly', alpha=0.8)
plt.xlabel('Fare Amount ($)')
plt.ylabel('Passenger Count')
plt.title('Isolation Forest: NYC Taxi Fare Anomaly Detection')
plt.legend()
plt.show()
```

## Question 8: Train a SARIMA model on the monthly airline passengers dataset. Forecast the next 12 months and visualize the results.
(Include your Python code and output in the code box below.)

```python
# Import libraries
import pandas as pd
import matplotlib.pyplot as plt
from statsmodels.tsa.statespace.sarimax import SARIMAX

# Load AirPassengers dataset from CSV (100% safe)
url = 'https://raw.githubusercontent.com/jbrownlee/Datasets/master/airline-passengers.csv'
data = pd.read_csv(url)

# Convert 'Month' to datetime and set as index
data['Month'] = pd.to_datetime(data['Month'])
data.set_index('Month', inplace=True)

# Ensure 'Passengers' is numeric
ts = data['Passengers'].astype(float)

# Set frequency to monthly
ts = ts.asfreq('MS')

# Plot original series
plt.figure(figsize=(12,4))
plt.plot(ts, label='Observed')
plt.title('AirPassengers Time Series')
plt.xlabel('Year')
plt.ylabel('Number of Passengers')
plt.show()

# Define SARIMA model: SARIMA(1,1,1)(1,1,1,12)
model = SARIMAX(ts,
                order=(1,1,1),
                seasonal_order=(1,1,1,12),
                enforce_stationarity=False,
                enforce_invertibility=False)

# Fit model
sarima_result = model.fit(disp=False)
print(sarima_result.summary())

# Forecast next 12 months
forecast = sarima_result.get_forecast(steps=12)
forecast_values = forecast.predicted_mean
forecast_ci = forecast.conf_int()

# Forecast index
forecast_index = pd.date_range(start=ts.index[-1] + pd.offsets.MonthBegin(1), periods=12, freq='MS')

# Plot forecast
plt.figure(figsize=(12,6))
plt.plot(ts, label='Observed', color='blue')
plt.plot(forecast_index, forecast_values, label='Forecast', color='orange')
plt.fill_between(forecast_index,
                 forecast_ci.iloc[:,0],
                 forecast_ci.iloc[:,1],
                 color='orange', alpha=0.3)
plt.title('SARIMA Forecast: AirPassengers Next 12 Months')
plt.xlabel('Year')
plt.ylabel('Number of Passengers')
plt.legend()
plt.show()
```

## Question 9: Apply Local Outlier Factor (LOF) on any numerical dataset to detect anomalies and visualize them using matplotlib.
(Include your Python code and output in the code box below.)

```python
# Import libraries
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
from sklearn.neighbors import LocalOutlierFactor

# Generate synthetic dataset
np.random.seed(42)

# Normal data points
X_normal = 0.5 * np.random.randn(200, 2)

# Add anomalies
X_anomalies = np.random.uniform(low=-4, high=4, size=(10, 2))

# Combine into one dataset
X = np.vstack([X_normal, X_anomalies])
data = pd.DataFrame(X, columns=['Feature1', 'Feature2'])

# Apply Local Outlier Factor
lof = LocalOutlierFactor(n_neighbors=20, contamination=0.05)
data['LOF_Label'] = lof.fit_predict(X)  # -1 for anomaly, 1 for normal
data['LOF_Score'] = lof.negative_outlier_factor_  # LOF scores

# Separate anomalies and normal points
anomalies = data[data['LOF_Label'] == -1]
normal = data[data['LOF_Label'] == 1]

# Plot results
plt.figure(figsize=(10,6))
plt.scatter(normal['Feature1'], normal['Feature2'], c='blue', label='Normal', alpha=0.5)
plt.scatter(anomalies['Feature1'], anomalies['Feature2'], c='red', label='Anomaly', alpha=0.8)
plt.xlabel('Feature1')
plt.ylabel('Feature2')
plt.title('Local Outlier Factor (LOF) Anomaly Detection')
plt.legend()
plt.show()
```
